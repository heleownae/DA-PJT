```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
import numpy as np
```

```python
# 타이타닉 데이터
titanic = pd.read_csv('C:/Users/LOVE/Downloads/vscode/ML/titanic/train.csv')

titanic.head(3)
```

```python
titanic.info()
```

```python
# 결측치 삭제하기
titanic.dropna(axis = 0).info()
```

```python
cond = (titanic['Age'].isna())
titanic[cond].head(5)
```

```python
cond2 = (titanic['Age'].notna())
titanic[cond2].info()
```

## fillna 이용한 대치

```python
age_mean = titanic[['Age']].mean().round(2)
titanic['Age_mean'] = titanic['Age'].fillna(age_mean)

titanic.info()
```

## SimpleImputer 이용한 대치

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer()

si.fit(titanic[['Age']])
```

```python
si.statistics_
```

```python
titanic['Age_si_mean'] = si.transform(titanic[['Age']])
titanic.info()
```
